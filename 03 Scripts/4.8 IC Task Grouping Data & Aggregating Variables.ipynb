#  Q4.  Create a loyalty flag for existing customers.

```python
# Create MAX order column
ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
# Confirm new column has been added
ords_prods_merge.head()
```

```python
# Create Loyalty flags
ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
# Check loyalty flags
ords_prods_merge['loyalty_flag'].value_counts(dropna = False)
```

```python
# Check loyalty flags for errors
ords_prods_merge[['user_id', 'loyalty_flag', 'order_number']].head(30)
```

#  Q6.  Create a spending flag for each user based on the average price across all their orders.

#### The team now wants to target different types of spenders in their marketing campaigns. This can be achieved by looking at the prices of the items people are buying. Create a spending flag for each user based on the average price across all their orders using the following criteria:  
####  If the mean of the prices of products purchased by a user is lower than 10, then flag them as a “Low spender.”  
####  If the mean of the prices of products purchased by a user is higher than or equal to 10, then flag them as a “High spender.”

```python
# Create "average spent" column
ords_prods_merge['average spent'] = ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
# Confirm new column has been added
ords_prods_merge.head()
```

```python
# Create flags for low and high spender
ords_prods_merge.loc[ords_prods_merge['average spent']<10, 'spender_flag']='Low_spender'
```

```python
ords_prods_merge.loc[ords_prods_merge['average spent']>= 10, 'spender_flag'] = 'High_spender'
```

```python
# Check spender_flags
ords_prods_merge['spender_flag'].value_counts(dropna = False)
```

```python
# Check spender flags for errors
ords_prods_merge[['user_id', 'average spent', 'spender_flag', 'prices']].head(100)
```

# Q7.  Determine frequent vs non-frequent customers

####  In order to send relevant notifications to users within the app (for instance, asking users if they want to buy the same item again), the Instacart team wants you to determine frequent versus non-frequent customers. Create an order frequency flag that marks the regularity of a user’s ordering behavior according to the median in the “days_since_prior_order” column. The criteria for the flag should be as follows:
####  If the median of “days_since_prior_order” is higher than 20, then the customer should be labeled a “Non-frequent customer.”
####  If the median is higher than 10 and lower than or equal to 20, then the customer should be labeled a “Regular customer.”
####  If the median is lower than or equal to 10, then the customer should be labeled a “Frequent customer.”

```python
# Create a column for frequent users
ords_prods_merge['Customer_orders_frequency'] = ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
# Confirm new column has been added
ords_prods_merge.head()
```

```python
#  Creating frequency flags
ords_prods_merge.loc[ords_prods_merge['Customer_orders_frequency'] > 20, 'frequency_flag'] = 'Non_frequent_customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['Customer_orders_frequency'] > 10) & (ords_prods_merge ['Customer_orders_frequency'] <=20),'frequency_flag'] = 'Regular_customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['Customer_orders_frequency'] <= 10, 'frequency_flag'] = 'Frequent_Customer'
```

```python
# Check frequency flags
ords_prods_merge['frequency_flag'].value_counts(dropna = False)
```

```python
# Check frequency flags for errors
ords_prods_merge[['user_id', 'days_since_prior_order', 'Customer_orders_frequency', 'frequency_flag']].head(100)
```
